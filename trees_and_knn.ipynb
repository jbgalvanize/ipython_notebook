##Import Needed Libraries

```python
import pandas as pd
import pydot
import os
import numpy as np
from sklearn import datasets
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import train_test_split, cross_val_score
from sklearn.externals.six import StringIO
from sklearn.neighbors import KNeighborsClassifier
```

#Read in Datasets we will use

```python
boston = pd.DataFrame(datasets.load_boston()['data'])
boston.columns = datasets.load_boston()['feature_names']
target = datasets.load_boston()['target']
#datasets.load_boston()['DESCR']

iris = pd.DataFrame(datasets.load_iris()['data'])
iris.columns = datasets.load_iris()['feature_names']
target_iris = datasets.load_iris()['target']
#datasets.load_iris()['DESCR']
```

#Fit Our Regression Tree to the Data

```python
trainX, testX, trainY, testY = train_test_split(boston.values, 
                                                target, test_size=.3, random_state=0)
```

```python
dt = DecisionTreeRegressor()
dt.fit(trainX, trainY)
pred1 = dt.predict(testX)
```

```python
dt2 = DecisionTreeRegressor(min_samples_leaf=30)
dt2.fit(trainX, trainY)
pred2 = dt2.predict(testX)
```

#How did they fit?

```python
np.mean(cross_val_score(dt, boston.values, target, cv = 5, scoring = 'mean_squared_error'))
```

```python
np.mean(cross_val_score(dt2, boston.values, target, cv = 5, scoring = 'mean_squared_error'))
```

#Now show a classification problem using Iris Dataset

```python
trainX_i, testX_i, trainY_i, testY_i = train_test_split(iris.values, target_iris, 
                                                test_size=.5, random_state=0)
```

```python
knn = KNeighborsClassifier() #default = 5
knn.fit(trainX_i, trainY_i)
iris_pred = knn.predict(testX_i)
```

```python
(testY_i == iris_pred).mean() 
```

```python
dt3 = DecisionTreeClassifier()
dt3.fit(trainX_i, trainY_i)
pred_i = dt3.predict(testX_i)
```

```python
(testY_i == pred_i).mean() 
```

```python
dt4 = DecisionTreeClassifier(min_samples_leaf = 10)
dt4.fit(trainX_i, trainY_i)
pred_i2 = dt4.predict(testX_i)
```

```python
(testY_i == pred_i2).mean() 
```

```python
dt5 = DecisionTreeClassifier(min_samples_leaf = 5)
dt5.fit(trainX_i, trainY_i)
pred_i3 = dt5.predict(testX_i)
```

```python
(testY_i == pred_i3).mean() 
```
